Tenemos 9888 entradas, podemos ver en el .info() que hay datos nulos, sobre todo en la columna de 'Manufacturer' 

Se va a proceder a limpiar la muestra de valores nulos para poder trabajar con una visualización clara y limpia.

También se va a separar la columna 'Price' para separar en dos: valor numérico y moneda.

```python
data.describe().T
```

```python
# Llamamos a la función y separamos en dos columnas aplicando función a toda la columna
data[['Precio', 'Currency']] = data['Price'].apply(sp.separa_valor_monedas).apply(pd.Series)
```

```python
# Voy a limpiar todas aquellas donde aparezca Â£ y cambiarlo por GBP
data.Currency.replace('Â£','GBP', inplace=True)
```

```python
data.Currency.value_counts()
```

```python
data['Precio_EUR'] = data.apply(sp.convert_currency, axis=1)
data.head()
```

* Ya tengo los datos que necesito, puedo deshacerme de la columna 'Price' y poner en su lugar 'Precio', 'Currency' y 'Precio_EUR'

```python
data = data.drop('Price', axis=1)
```

```python
data.head()
```

```python
data.columns
```

```python
# Voy a reordenar las columnas como me apetece:
data = data[['Precio','Currency', 'Precio_EUR', 'Boat Type', 'Manufacturer', 'Type', 'Year Built', 'Length', 'Width',
       'Material', 'Location', 'Number of views last 7 days']]
data.head()
```

```python
# Ahora voy a cambiar los NaNs de la columna material con 'unknown'
data.Material.fillna('unknown', inplace=True)
```

```python
data.info()
```

```python
# Y procedo igual con manufacturer:
data.Manufacturer.fillna('unknown', inplace=True)
data.info()
```

```python
# El resto de valores nulos que haya los voy a borrar porque es info importante
data.dropna(inplace=True)
data.info()
```

```python
data.head(10)
```

Ahora voy a pasar a limpiar la columna 'Location' porque parece hay varios caracteres feos
* Lo voy a hacer de manera local y no con una función externa, solo aplica a este caso.

```python
data[['Country', 'City']] = data.Location.str.split('Â»', n=1, expand=True) # Separo por 'Â»' por verlo en el head() como separa País y ciudades

data[['City', 'Other']] = data.City.str.split('Â»', n=1, expand=True) # Separo la nueva 'City' en otros -> a veces hay estado o mas detalles de localización

data.City = data.City.str.replace('©', '', regex=True) # Limpio también caracteres raros
data.City = data.City.str.replace('¶', '', regex=True)
data.City = data.City.str.replace('Ã', '', regex=True)

data.head(10) # A ver como ha quedado
```

```python
# Una vez separado en tres columnas, a simple vista puedo ver que la columna 'Other' no va a aportar gra información y puede traer problemas 
# así que voy a deshacerme de ella, así como quitar la columna Location y ordenar a mi gusto.

data = data.drop(['Location', 'Other'], axis=1)
```

```python
data.columns
```

```python
data = data[['Precio', 'Currency', 'Precio_EUR', 'Boat Type', 'Manufacturer', 'Type',
       'Year Built', 'Length', 'Width', 'Material', 'Country', 'City', 'Number of views last 7 days']]
data.head()
```

De cara a tener una base de datos 'friendly' con PowerBI a cambiar los separadores de decimales de '.' a ',' para las oclumnas Precio_EUR, Length y Width

```python
import pandas as pd

# Assuming 'Precio_EUR', 'Length', and 'Width' are columns in your DataFrame
data['Precio_EUR'] = data['Precio_EUR'].astype(str).apply(pd.Series.str.replace, pat='.', repl=',')
data['Length'] = data['Length'].astype(str).apply(pd.Series.str.replace, pat='.', repl=',')
data['Width'] = data['Width'].astype(str).apply(pd.Series.str.replace, pat='.', repl=',')

data.head()

data.head()
```

```python
# Listo, me quedo con esta base de datos. La paso a mi carpeta de data como boat_data_clean
data.to_csv('../data/boat_data_clean.csv')
```
